### Αλγόριθμος παραγωγής λαβυρίνθων
```
1. Αρχικά όλα τα κελιά περιέχουν εμπόδια και η λίστα των τοίχων είναι κενή.
2. Ένα κελί επιλέγεται ως μέρος του λαβυρίνθου. Οι τοίχοι του κελιού προστίθενται στη λίστα των τοίχων.
3. Όσο η λίστα των τοίχων δεν είναι κενή:
      1. Επιλέγεται τυχαία ένας τοίχος από τη λίστα. Εαν μόνο ένα από τα (δύο) κελιά που χωρίζει ο τοίχος είναι μέρος του λαβυρίνθου, τότε:
            1. Ο τοίχος και το κελί που δεν είναι μέρος του λαβυρίνθου, γίνονται μέρος του λαβυρίνθου.
            2. Οι τοίχοι των νέων κελιών που προστέθηκαν στο λαβύρινθο προστίθενται στη λίστα των τοίχων.
      2. Ο τοίχος αφαιρείται από τη λίστα των τοίχων
```
*ως τοίχοι ενός κελιού θεωρούνται όσα απο τα 4 γειτονικά κελιά (όχι τα διαγώνια) περιέχουν εμπόδιο

# Κώδικας visualization


````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
from IPython.display import HTML

class visualization:
    def __init__(self, S, F):
        '''
          Η μέθοδος αυτή αρχικοποιεί ένα αντικείμενο τύπου visualization.
          Είσοδος: 
          -> S: το σημείο εκκίνσης της αναζήτησης
          -> F: το σημείο τερματισμού
        '''
        self.S = S
        self.F = F
        self.images = []
    
    def draw_step(self, grid, frontier, expanded_nodes):
        '''
          Η συνάρτηση αυτή καλείται για να σχεδιαστεί ένα frame στο animation (πρακτικά έπειτα από την επέκταση κάθε κόμβου)
          Είσοδος: 
          -> grid: Ένα χάρτης τύπου grid
          -> frontier: Μια λίστα με τους κόμβους που ανήκουν στο μέτωπο της αναζήτησης
          -> expanded_nodes: Μια λίστα με τους κόμβους που έχουν ήδη επεκταθεί
          Επιστρέφει: None
          Η συνάρτηση αυτή πρέπει να καλεστεί τουλάχιστον μια φορά για να μπορέσει να σχεδιαστει ένα animation (πρεπεί το animation να έχει τουλάχιστον ένα frame).
        '''
        image = np.zeros((grid.N, grid.N, 3), dtype=int)
        image[grid.grid == 0] = [255, 255, 255]
        image[grid.grid == 1] = [0, 0, 0]
        
        for node in expanded_nodes:
            image[node] = [0, 0, 128]

        for node in frontier:
            image[node] = [0, 225, 0]

        image[self.S] = [50, 168, 64]
        image[self.F] = [168, 50, 50]
        self.images.append(image)
    
    def add_path(self, path):
        '''
          Η συνάρτηση αυτή προσθέτει στο τελευταίο frame το βέλτιστο μονοπάτι.
          Είσοδος:
          -> path: Μια λίστα η όποια περιέχει το βέλτιστο μονοπάτι (η οποία πρέπει να περιέχει και τον κόμβο αρχή και τον κόμβο στόχο)
          Έξοδος: None
        '''
        for n in path[1:-1]:
            image = np.copy(self.images[-1])
            image[n] = [66, 221, 245]
            self.images.append(image)
        for _ in range (100):
            self.images.append(image)
        
    def create_gif(self, fps = 30, repeat_delay = 2000):
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        fig = plt.figure()
        plt.axis('off')
        ims = []
        for img in self.images:
            img = plt.imshow(img)
            ims.append([img])
        ani = animation.ArtistAnimation(fig, ims, interval=1000//fps, blit=True, repeat_delay= repeat_delay)
        plt.close(fig)
        return ani
        
    def save_gif(self, filename, fps = 30):
        '''
            Η συνάρτηση αυτή ξαναδημιουργεί και αποθηκεύει το animation σε ένα αρχείο.
            Είσοδος:
            -> Το όνομα του αρχείου με κατάληξη .gif
            Έξοδος: (None)
        '''
        ani = self.create_gif(fps)
        writer = PillowWriter(fps= fps)
        ani.save(filename, writer=writer)

    def show_gif(self, fps= 30, repeat_delay = 2000):
        '''
            Η συνάρτηση αυτή εμφανίζει inline το animation.
            Είσοδος:
            -> fps: τα frames per second
            Έξοδος: Το αντικείμενο που παίζει το animation
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        print(len(self.images))
        ani = self.create_gif(fps, repeat_delay)
        # return HTML(ani.to_html5_video())
        return HTML(ani.to_jshtml())

    def show_last_frame(self):
        '''
            Η μέθοδος αυτή εμφανίζει inline το τελευταίο frame που έχει δημιουργήθει.
            Είσοδος:
            Έξοδος: Το αντικείμενο που εμφανίζει την εικόνα.
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        else:
            plt.imshow(self.images[-1])


class EmptyStackOfImages(Exception):
    pass
````

###<b>Κατασκευή Λαβυρίνθων</b>
***
Στο μέρος αυτό κληθήκαμε να υλοποιήσουμε την παρακάτω κλάση, η οποία παράγει λαβυρίνθους μεγέθους N, με αρχικό σημείο το S και τελικό το F. Ο αλγόριθμος κατασκευής που χρησιμοποιήσαμε είναι αυτός που φαίνεται στην εκφώνηση. Ξεκινώντας από κάθε κελί, προσθέτουμε στη λίστα cells τους γείτονες του κελιού που εξετάζουμε, εφόσον το κελί αυτό έχει λιγότερους από 2 γείτονες (για την αποφυγή δημιουργίας γειτόνων). <br><br>
Παρατηρήσαμε ότι ο αλγόριθμος αυτός, σε συνδυασμό με το "καθάρισμα" αρχικά των σημείων S, F επαρκεί ώστε σε κάθε λαβύρινθο που κατασκευάζεται να υπάρχει μονοπάτι από το S στο F. Ωστόσο, πολλές φορές το μονοπάτι αυτό είναι μοναδικό, γεγονός που θα παίξει ρόλο στα αποτελέσματα του τελευταίου μέρους.

````python
%matplotlib inline
import numpy as np
from queue import LifoQueue
from random import shuffle, uniform
import random
import matplotlib.pyplot as plt


class grid:
    def __init__(self, N, S, F):
        
        ## Make sure start and end are within the grid
        assert N > 2
        assert S[0] < N
        assert S[1] < N
        assert F[0] < N
        assert F[1] < N

        assert S[0] > 0
        assert S[1] > 0
        assert F[0] > 0
        assert F[1] > 0

        self.N = N

        ## Initialize grid with obstacles
        self.grid = np.ones((N, N), dtype=np.int32)
        
        ## Start and end position have no obstacles
        self.grid[S]=0
        self.grid[F]=0
        explored = {S}

        ## OUR CODE BEGINS HERE
        
        cells = [(S[0]+1, S[1]), (S[0], S[1]+1), (S[0]-1, S[1]), (S[0], S[1]-1)]

        while (cells):
          cell = random.choice(cells)
          neighbors = {(cell[0]+1, cell[1]), (cell[0]-1, cell[1]), (cell[0], cell[1]+1), (cell[0], cell[1]-1)}

          if (0<cell[0]<N and 0<cell[1]<N and len(explored & neighbors) < 2):
            # Clear cell
            self.grid[cell] = 0

            # Add cell's neighbors to list
            if ((cell[0]+1, cell[1]) not in explored):
              cells.append((cell[0]+1, cell[1]))
            if ((cell[0]-1, cell[1]) not in explored):
              cells.append((cell[0]-1, cell[1]))
            if ((cell[0], cell[1]+1) not in explored):
              cells.append((cell[0], cell[1]+1))
            if ((cell[0], cell[1]-1) not in explored):
              cells.append((cell[0], cell[1]-1))
            
            # Add cell to explored set
            explored.add(cell)
          
          cells.remove(cell)
        


    def draw_map(self, S=None, F=None, path=None):
        image = np.zeros((self.N, self.N, 3), dtype=int)
        image[self.grid == 0] = [255, 255, 255]
        image[self.grid == 1] = [0, 0, 0]
        if S:
            image[S] = [50, 168, 64]
        if F:
            image[F] = [168, 50, 50]
        if path:
            for n in path[1:-1]:
                image[n] = [66, 221, 245]

        plt.imshow(image)
        plt.xticks([])
        plt.yticks([])
        plt.show()
````

Στο μέρος αυτό κληθήκαμε να υλοποιήσουμε διάφορους αλγορίθμους αναζήτησης μονοπατιών μεταξύ δύο κόμβων του λαβυρίνθου. Οι αλγόριθμοι που κληθήκαμε να υλοποιήσουμε διέφεραν ως προς την πραγματική απόσταση μεταξύ των κόμβων και την ευριστική συνάρτηση. Συγκεκριμένα εξετάσαμε τις εξής περιπτώσεις: <br>
>Πραγματική Απόσταση | Ευριστική Συνάρτηση
>--- | ---
> 0 | {Manhattan, Euclidean}
> 1 | 0
> 1 | {Manhattan, Euclidean}
<br>

Παρατηρήσαμε ότι η **πρώτη περίπτωση** είναι στην ουσία ο αλγόριθμος **Best First**, ο οποίος δεν μας εγγυάται ότι θα βρει τη βέλτιστη λύση. Από τη **δεύτερη περίπτωση** προκύπτει ο αλγόριθμος του **Dijkstra**, ο οποίος ούτε αυτός μας εγγυάται την εύρεση του βέλτιστου μονοπατιού. Τέλος, **τρίτη περίπτωση** είναι ο αλγόριθμος **Α***. Δεδομένου του ότι η απόσταση του εκάστοτε κόμβου (είτε ευκλίδεια είναι Manhattan) μπορεί να γίνει μεγαλύτερη από 1, τότε ούτε ο Α* σε αυτή τη περίπτωση δεν βρίσκει σίγουρα το βέλτιστο μονοπάτι.

````python
class pathfinder:
    i = 1
    def __init__(self, S, F, grid, c, h):
        self.S = S
        self.F = F
        self.grid = grid
        self.cost = c
        self.heuristic = h
        #for visualization 
        self.vis = visualization(S, F)
        self.path=[]


    def find_path(self):
        frontier = []
        expanded_nodes = {(self.S[0], self.S[1], self.S[0], self.S[1])}
        expanded = []
        front = []
        current_node = self.S

        while ((current_node[0],current_node[1]) != (self.F[0],self.F[1])):
          n = current_node

          if (0<n[0]-1<N and 0<n[1]<N and self.grid.grid[(n[0]-1,n[1])] == 0 and (n[0]-1,n[1],n[0],n[1]) not in expanded_nodes and (n[0],n[1],n[0]-1,n[1]) not in expanded_nodes):
            cost = self.cost(n[0]-1,n[1]) + self.heuristic(n[0]-1,n[1])
            frontier.append((n[0]-1, n[1], cost, n[0], n[1]))
            front.append((n[0]-1, n[1]))
          if (0<n[0]+1<N and 0<n[1]<N and self.grid.grid[(n[0]+1,n[1])] == 0 and (n[0]+1,n[1],n[0],n[1]) not in expanded_nodes and (n[0],n[1],n[0]+1,n[1]) not in expanded_nodes):
            cost = self.cost(n[0]+1,n[1]) + self.heuristic(n[0]+1,n[1])
            frontier.append((n[0]+1, n[1], cost, n[0], n[1]))
            front.append((n[0]+1, n[1]))
          if (0<n[0]<N and 0<n[1]-1<N and self.grid.grid[(n[0],n[1]-1)] == 0 and (n[0],n[1]-1,n[0],n[1]) not in expanded_nodes and (n[0],n[1],n[0],n[1]-1) not in expanded_nodes):
            cost = self.cost(n[0],n[1]-1) + self.heuristic(n[0],n[1]-1)
            frontier.append((n[0], n[1]-1, cost, n[0], n[1]))
            front.append((n[0], n[1]-1))
          if (0<n[0]<N and 0<n[1]+1<N and self.grid.grid[(n[0],n[1]+1)] == 0 and (n[0],n[1]+1,n[0],n[1]) not in expanded_nodes and (n[0],n[1],n[0],n[1]+1) not in expanded_nodes):
            cost = self.cost(n[0],n[1]+1) + self.heuristic(n[0],n[1]+1)
            frontier.append((n[0], n[1]+1, cost, n[0], n[1]))
            front.append((n[0], n[1]+1))

          # Sort list by cost
          frontier.sort(key=lambda x: x[2])

          #Explore best node
          current_node = frontier.pop(0)
          current_node = (current_node[0], current_node[1], current_node[3], current_node[4])
          expanded_nodes.add(current_node)

          # Uncomment for visualization
          '''
          for f in front:
            if (f[0]==current_node[0] and f[1]==current_node[1]):
              front.remove(f)

          expanded.append((current_node[0], current_node[1]))
          self.vis.draw_step(self.grid, front, expanded)
          '''

        # Reconstruct path
        n = current_node
        while (n[0]!=n[2] or n[1]!=n[3]):
          # Add current node to path
          self.path.append((n[0],n[1]))

          # Move back to the source of current node
          for m in expanded_nodes:
            if (m[0]==n[2] and m[1]==n[3]):
              n = m
              break

        # Finally add Starting node to path
        self.path.append((self.S[0],self.S[1]))

        # Uncomment for visualization
        '''
        self.vis.add_path(self.path)
        # self.vis.show_gif(fps = 15)
        self.vis.save_gif("gif%d.gif" %(pathfinder.i), fps=15)
        pathfinder.i += 1
        '''
        
    def get_path(self):
        return self.path

````

 - Ομοιόμορφη Κατανομή
 - Τριγωνική Κατανομή
 - Κατανομή Βήτα

 Συγκεκριμένα, ορίσαμε τη μέθοδο assign_costs στην κλάση grid, η οποία παίρνει σαν όρισμα μία γεννήτρια τυχαίων αριθμών και αναθέτει σε κάθε κελί του λαβυρίνθου μία τυχαία τιμή κόστους. Επίσης, από την κλάση pathfinder αφαιρέσαμε το όρισμα για το κόστος της πραγματικής απόστασης μεταξύ δύο κελιών, καθώς αυτή η πληροφορία πλέον είναι αποθηκευμένη στο ίδιο το grid και τροποποιήσαμε αντίστοιχα τον κώδικα. Προσθέσαμε επίσης getters για τα στοιχεία που θα χρειαστούμε στο τελευταίο κομμάτι του τρίτου μέρους, δηλαδή του μονοπατιού (που υπήρχε ήδη) και των επεκτεταμμένων κόμβων κατά την αναζήτηση του.

#### Η τροποποιημένη κλάση grid.

````python
%matplotlib inline
import numpy as np
from queue import LifoQueue
from random import shuffle, uniform
import random
import matplotlib.pyplot as plt


class grid:
    def __init__(self, N, S, F):
        
        ## Make sure start and end are within the grid
        assert N > 2
        assert S[0] < N
        assert S[1] < N
        assert F[0] < N
        assert F[1] < N

        assert S[0] > 0
        assert S[1] > 0
        assert F[0] > 0
        assert F[1] > 0

        self.N = N

        ## Initialize grid with obstacles
        self.grid = np.ones((N, N), dtype=np.int32)
        self.cost = np.ones((N, N), dtype=float)
        
        ## Start and end position have no obstacles
        self.grid[S]=0
        self.grid[F]=0
        explored = {S}

        ## YOUR CODE HERE
        
        cells = [(S[0]+1, S[1]), (S[0], S[1]+1), (S[0]-1, S[1]), (S[0], S[1]-1)]

        while (cells):
          cell = random.choice(cells)
          neighbors = {(cell[0]+1, cell[1]), (cell[0]-1, cell[1]), (cell[0], cell[1]+1), (cell[0], cell[1]-1)}

          if (0<cell[0]<N and 0<cell[1]<N and len(explored & neighbors) < 2):
            # Clear cell
            self.grid[cell] = 0

            # Add cell's neighbors to list
            if ((cell[0]+1, cell[1]) not in explored):
              cells.append((cell[0]+1, cell[1]))
            if ((cell[0]-1, cell[1]) not in explored):
              cells.append((cell[0]-1, cell[1]))
            if ((cell[0], cell[1]+1) not in explored):
              cells.append((cell[0], cell[1]+1))
            if ((cell[0], cell[1]-1) not in explored):
              cells.append((cell[0], cell[1]-1))
            
            # Add cell to explored set
            explored.add(cell)
          
          cells.remove(cell)
        


    def draw_map(self, S=None, F=None, path=None):
        image = np.zeros((self.N, self.N, 3), dtype=int)
        image[self.grid == 0] = [255, 255, 255]
        image[self.grid == 1] = [0, 0, 0]
        if S:
            image[S] = [50, 168, 64]
        if F:
            image[F] = [168, 50, 50]
        if path:
            for n in path[1:-1]:
                image[n] = [66, 221, 245]

        plt.imshow(image)
        plt.xticks([])
        plt.yticks([])
        plt.show()



    # Assign random costs to clear cells
    def assign_costs(self, r):
        for i in range(N):
          for j in range(N):
            if(self.grid[i][j] == 0):
              self.cost[i][j] = r()
````

#### Η τροποποιημένη κλάση pathfinder.

````python
class pathfinder:
    def __init__(self, S, F, grid, h):
        self.S = S
        self.F = F
        self.grid = grid
        self.heuristic = h
        #for visualization 
        self.vis = visualization(S, F)
        self.path=[]
        self.expanded_nodes = {}


    def find_path(self):
        frontier = []
        self.expanded_nodes = {(self.S[0], self.S[1], self.S[0], self.S[1])}
        current_node = self.S

        while ((current_node[0],current_node[1]) != (self.F[0],self.F[1])):
          n = current_node

          if (0<n[0]-1<N and 0<n[1]<N and self.grid.grid[(n[0]-1,n[1])] == 0 and (n[0]-1,n[1],n[0],n[1]) not in self.expanded_nodes and (n[0],n[1],n[0]-1,n[1]) not in self.expanded_nodes):
            cost = self.grid.cost[n[0]-1,n[1]] + self.heuristic(n[0]-1,n[1])
            frontier.append((n[0]-1, n[1], cost, n[0], n[1]))
          if (0<n[0]+1<N and 0<n[1]<N and self.grid.grid[(n[0]+1,n[1])] == 0 and (n[0]+1,n[1],n[0],n[1]) not in self.expanded_nodes and (n[0],n[1],n[0]+1,n[1]) not in self.expanded_nodes):
            cost = self.grid.cost[n[0]+1,n[1]] + self.heuristic(n[0]+1,n[1])
            frontier.append((n[0]+1, n[1], cost, n[0], n[1]))
          if (0<n[0]<N and 0<n[1]-1<N and self.grid.grid[(n[0],n[1]-1)] == 0 and (n[0],n[1]-1,n[0],n[1]) not in self.expanded_nodes and (n[0],n[1],n[0],n[1]-1) not in self.expanded_nodes):
            cost = self.grid.cost[n[0],n[1]-1] + self.heuristic(n[0],n[1]-1)
            frontier.append((n[0], n[1]-1, cost, n[0], n[1]))
          if (0<n[0]<N and 0<n[1]+1<N and self.grid.grid[(n[0],n[1]+1)] == 0 and (n[0],n[1]+1,n[0],n[1]) not in self.expanded_nodes and (n[0],n[1],n[0],n[1]+1) not in self.expanded_nodes):
            cost = self.grid.cost[n[0],n[1]+1] + self.heuristic(n[0],n[1]+1)
            frontier.append((n[0], n[1]+1, cost, n[0], n[1]))

          # Sort list by cost
          frontier.sort(key=lambda x: x[2])

          #Explore best node
          current_node = frontier.pop(0)
          current_node = (current_node[0], current_node[1], current_node[3], current_node[4])
          self.expanded_nodes.add(current_node)

          # Uncomment for visualization
          '''
          for f in front:
            if (f[0]==current_node[0] and f[1]==current_node[1]):
              front.remove(f)

          expanded.append((current_node[0], current_node[1]))
          self.vis.draw_step(self.grid, front, expanded)
          '''

        # Reconstruct path
        n = current_node
        while (n[0]!=n[2] or n[1]!=n[3]):
          # Add current node to path
          self.path.append((n[0],n[1]))

          # Move back to the source of current node
          for m in self.expanded_nodes:
            if (m[0]==n[2] and m[1]==n[3]):
              n = m
              break

        # Finally add Starting node to path
        self.path.append((self.S[0],self.S[1]))

        # Uncomment for visualization
        '''
        self.vis.add_path(self.path)
        # self.vis.show_gif(fps = 15)
        self.vis.save_gif("gif%d.gif" %(pathfinder.i), fps=15)
        pathfinder.i += 1
        '''
        
    def get_path(self):
        return self.path
    
    def get_expanded_nodes_length(self):
        return len(self.expanded_nodes)

````

#### Παραγωγή Γραφημάτων
***
Στο παρακάτω κελί κώδικα δημιουργούμε λαβυρίνθους με μεγέθη ${[5, 10, 15, ..., 100]}$ και για καθένα από αυτούς:
1. Αναθέσαμε τυχαία κόστη στα κελιά του σύμφωνα με τις κατανομές που αναφέραμε παραπάνω.
2. Αναζητήσαμε το βέλτιστο μονοπάτι χρησιμοποιώντας καθεμία από τις τρεις διαφορετικές περιπτώσεις ευριστικών συναρτήσεων.
3. Αποτυπώσαμε το μήκος του βέλτιστου μονοπατιού και το πλήθος των επεκτεταμμένων κόμβων της κάθε αναζήτησης για κάθε περίπτωση σε δύο γραφικές παραστάσεις.

````python
import random
from functools import partial
from scipy.spatial import distance
import matplotlib.pyplot as plt

S = (5, 9)
F = (45, 35)

# Three random number generators between (0,1)
random_generators = [
    partial(random.uniform, 0, 1),
    partial(random.triangular, 0, 1, 0.5),
    partial(random.betavariate, 1, 1)
]

# Three heuristic functions
heuristics = [
    lambda x, y: 0,
    lambda x, y: distance.cityblock([S[0],S[1]], [x,y]),
    lambda x, y: distance.euclidean([S[0],S[1]], [x,y])
]

# The sizes of the mazes
maze_sizes = []
maze_sizes.extend(range(5, 105, 5))

# Creating random generators - heuristics combinations
random_generators_names = ['Uniform', 'Triangular', 'Betavariate']
heuristics_names = ['Zero', 'Manhattan', 'Euclidean']
combos = []

for r in random_generators_names:
  for h in heuristics_names:
    combos.append(r + " - " + h)

i = 0
path_lengths = [[], [], [], [], [], [], [], [], []]
expanded_nodes_lengths = [[], [], [], [], [], [], [], [], []]

# Generate a single maze for each size
for N in maze_sizes:
  S = (1,2)
  F = (N-3, N-2)
  map = grid(N, S, F)
  # map.draw_map(S, F)

  # Iterate through all random generator - heuristic combinations
  for r in random_generators:
    for h in heuristics:
      map.assign_costs(r)
      pf = pathfinder(S, F, map, h)
      pf.find_path()
      # map.draw_map(S, F, pf.get_path())
      path_lengths[i%9].append(len(pf.get_path()))
      expanded_nodes_lengths[i%9].append(pf.get_expanded_nodes_length())
      i += 1

fig = plt.figure(figsize=(15, 15))

ax1 = fig.add_subplot(211)

for l in path_lengths:
  ax1.plot(maze_sizes, l)
ax1.legend(combos, loc='upper left')
ax1.set_xlabel("Maze size")
ax1.set_ylabel("Best path length")
ax1.set_title("Path length for different Distributions and Heuristics combinations")

ax2 = fig.add_subplot(212)

for l in expanded_nodes_lengths:
  ax2.plot(maze_sizes, l)
ax2.legend(combos, loc='upper left')
ax2.set_xlabel("Maze size")
ax2.set_ylabel("Expanded Nodes")
ax2.set_title("Expanded nodes for different Distributions and Heuristics combinations")
````
